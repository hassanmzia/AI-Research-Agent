````python
class ResearchPhase(Enum):
    """Phases of the research workflow."""
    INITIALIZATION = "initialization"
    PLANNING = "planning"
    DISCOVERY = "discovery"
    EVALUATION = "evaluation"
    COMPLETION = "completion"

class AgentRole(Enum):
    """Roles for different agents in the system."""
    LEAD_SUPERVISOR = "lead_supervisor"
    PLANNER = "planner"
    DISCOVERY_SUPERVISOR = "discovery_supervisor"
    EVALUATION_SUPERVISOR = "evaluation_supervisor"

class ResearchSystemState(TypedDict):
    """Overall system state for the research pipeline."""
    # Request information
    request_id: str
    research_objective: str

    # Phase tracking
    current_phase: ResearchPhase
    phase_history: List[Dict[str, Any]]

    # Data storage
    discovered_papers: List[Dict[str, Any]]
    analysis_results: List[Dict[str, Any]]
    evaluation_results: List[Dict[str, Any]]
    synthesis_data: Optional[Dict[str, Any]]
    final_report: Optional[str]

    # Agent communication
    messages: List[BaseMessage]
    pending_tasks: List[Dict[str, Any]]
    completed_tasks: List[Dict[str, Any]]

    # Error tracking
    errors: List[Dict[str, Any]]
    warnings: List[str]

    # Metadata
    created_at: Optional[datetime]
    updated_at: Optional[datetime]
    total_papers_processed: int
    total_processing_time: float
````

### Log Utils

````python
class ResearchLogger:
    """Simplified logging utility for the research system."""

    def __init__(self, name: str):
        self.name = name
        self.logger = logging.getLogger(name)
        if not self.logger.handlers:
            handler = logging.StreamHandler()
            formatter = logging.Formatter(
                '%(asctime)s - %(name)s - %(levelname)s - %(message)s'
            )
            handler.setFormatter(formatter)
            self.logger.addHandler(handler)
            self.logger.setLevel(logging.INFO)

    def info(self, message: str, **kwargs) -> None:
        """Log info message with optional context."""
        if kwargs:
            message = f"{message} | {kwargs}"
        self.logger.info(message)

    def warning(self, message: str, **kwargs) -> None:
        """Log warning message with optional context."""
        if kwargs:
            message = f"{message} | {kwargs}"
        self.logger.warning(message)

    def error(self, message: str, **kwargs) -> None:
        """Log error message with optional context."""
        if kwargs:
            message = f"{message} | {kwargs}"
        self.logger.error(message)

def get_logger(name: str) -> ResearchLogger:
    """Get a logger instance."""
    return ResearchLogger(name)
````

### Retry for LLM


````python
@retry(
    stop=stop_after_attempt(3),
    wait=wait_exponential(multiplier=1, min=4, max=60),
    retry=retry_if_exception_type((APIError, RateLimitError, APITimeoutError))
)
def call_llm_with_retry(llm, messages, logger):
    """Call LLM with automatic retry on transient failures"""
    try:
        return llm.invoke(messages)
    except RateLimitError as e:
        logger.warning(f"Rate limit hit, backing off: {e}")
        raise  # Will be retried by @retry decorator
    except APITimeoutError as e:
        logger.warning(f"API timeout, retrying: {e}")
        raise
````

````python

def search_arxiv(query: str, max_papers: int = 10, from_date: str = None, to_date: str = None) -> List[Dict[str, Any]]:
    """
    Search arXiv for research papers matching the given query with date range filtering.

    Args:
        query: Search query for papers (e.g., "artificial general intelligence")
        max_papers: Maximum number of papers to retrieve (default: 20)
        from_date: Start date for papers (YYYY-MM-DD format, e.g., "2025-08-01")
        to_date: End date for papers (YYYY-MM-DD format, e.g., "2025-08-18")

    Returns:
        List of paper dictionaries with metadata including title, authors, abstract, URL, etc.
    """
    logger = get_logger("search_arxiv")
    logger.info(f"Starting ArXiv search",
               query=query, max_papers=max_papers, from_date=from_date, to_date=to_date)


    ## You can Hard code From_date and to_date for consistent results to demonstrate the system
    ## Uncomment out the below hard coded dates and modify them as necessary.
    # from_date = "2025-06-01"
    # to_date = "2025-06-14"

    try:
        # Build arXiv search query with category filtering
        categories = ["cs.AI", "cs.LG", "cs.CL", "cs.CV", "cs.NE"]
        category_filter = " OR ".join([f"cat:{cat}" for cat in categories])

        # Build the full query starting with content and categories
        full_query = f'({query}) AND ({category_filter})'

        # Add date range filter if provided
        if from_date or to_date:
            # Convert dates to arXiv API format (YYYYMMDDHHMM)
            if from_date:
                from_date_arxiv = datetime.strptime(from_date, "%Y-%m-%d").strftime("%Y%m%d0000")
            else:
                from_date_arxiv = "202001010000"  # Default to 2020

            if to_date:
                to_date_arxiv = datetime.strptime(to_date, "%Y-%m-%d").strftime("%Y%m%d2359")
            else:
                to_date_arxiv = datetime.now().strftime("%Y%m%d2359")  # Default to today

            # Use space instead of + for the TO operator
            date_filter = f"submittedDate:[{from_date_arxiv} TO {to_date_arxiv}]"
            full_query = f'{full_query} AND {date_filter}'
            logger.info(f"Date filter applied: {date_filter}")

        logger.info(f"Full ArXiv query: {full_query}")

        # Create search client
        search = arxiv.Search(
            query=full_query,
            max_results=max_papers,
            sort_by=arxiv.SortCriterion.SubmittedDate,
            sort_order=arxiv.SortOrder.Descending
        )

        # Execute search
        results = list(search.results())
        papers = []

        # Process results
        for result in results:
            # Extract paper metadata in the required format
            paper = {
                "id": result.get_short_id(),
                "title": result.title,
                "link": result.entry_id,
                "metadata": {
                    "authors": [
                        {
                            "name": author.name,
                            "affiliation": ""  # arXiv doesn't provide affiliation
                        } for author in result.authors
                    ],
                    "abstract": result.summary.replace('\n', ' ').strip(),
                    "published_date": result.published.isoformat(),
                    "updated_date": result.updated.isoformat(),
                    "categories": result.categories,
                    "source": "arxiv",
                    "doi": getattr(result, 'doi', None),
                    "comment": getattr(result, 'comment', None),
                    "journal_ref": getattr(result, 'journal_ref', None)
                }
            }
            papers.append(paper)

        logger.info(f"ArXiv search completed", papers_found=len(papers))
        return papers

    except Exception as e:
        logger.error(f"ArXiv search failed: {e}")
        return []

@tool
def discover_and_process_papers(query: str, max_papers: int = 10, from_date: str = None, to_date: str = None) -> Dict[str, Any]:
    """
    Complete discovery workflow tool that searches, deduplicates, and validates papers.

    Args:
        query: Search query for papers
        max_papers: Maximum total papers to retrieve (default: 30)
        from_date: Start date (YYYY-MM-DD format, optional)
        to_date: End date (YYYY-MM-DD format, optional)

    Returns:
        Dictionary with processed_papers, statistics, source_counts, and search_metadata
    """
    logger = get_logger("discover_and_process")
    logger.info(f"Starting complete discovery workflow",
               query=query, max_papers=max_papers, from_date=from_date, to_date=to_date)

    start_time = time.time()

    try:
        # Step 1: Search ArXiv
        papers = search_arxiv(query, max_papers, from_date, to_date)

        # Step 2: Basic deduplication by title similarity
        unique_papers = []
        seen_titles = set()

        for paper in papers:
            title = paper['title'].lower().strip()
            # Simple title normalization for deduplication
            normalized_title = re.sub(r'\s+', ' ', title)
            if normalized_title not in seen_titles:
                seen_titles.add(normalized_title)
                unique_papers.append(paper)

        # Step 3: Validation - ensure papers have required fields
        valid_papers = []
        for paper in unique_papers:
            if (paper.get('title') and
                paper.get('metadata', {}).get('abstract') and
                len(paper.get('metadata', {}).get('abstract', '')) >= 50):
                valid_papers.append(paper)

        # Step 4: Generate statistics
        processing_time = time.time() - start_time
        statistics = {
            "initial_count": len(papers),
            "after_deduplication": len(unique_papers),
            "final_count": len(valid_papers),
            "duplicates_removed": len(papers) - len(unique_papers),
            "invalid_removed": len(unique_papers) - len(valid_papers),
            "processing_time": f"{processing_time:.2f}s"
        }

        source_counts = {
            "arxiv": len(valid_papers),
            "ieee": 0,
            "acl": 0
        }

        search_metadata = {
            "query_used": query,
            "date_range": f"{from_date} to {to_date}" if from_date or to_date else "all_time",
            "sources_searched": ["arxiv"],
            "processing_timestamp": datetime.now().isoformat(),
            "processing_time": f"{processing_time:.2f}s"
        }

        result = {
            "processed_papers": valid_papers,
            "statistics": statistics,
            "source_counts": source_counts,
            "search_metadata": search_metadata
        }

        logger.info(f"Discovery workflow completed",
                   initial_papers=len(papers),
                   final_papers=len(valid_papers),
                   processing_time=f"{processing_time:.2f}s")

        return result

    except Exception as e:
        logger.error(f"Discovery workflow failed: {e}")
        return {
            "processed_papers": [],
            "statistics": {"error": str(e)},
            "source_counts": {"arxiv": 0, "ieee": 0, "acl": 0},
            "search_metadata": {"error": str(e)}
        }

# Discovery tools list for LangGraph agents
DISCOVERY_TOOLS = [discover_and_process_papers]
````

````python
def get_agi_metrics_prompt(title: str, abstract: str, authors: List[str]) -> str:
    """
    Generate the AGI evaluation prompt for a research paper.

    Args:
        title: Paper title
        abstract: Paper abstract
        authors: List of author names

    Returns:
        Formatted prompt string for AGI evaluation
    """
    authors_str = ", ".join(authors[:5])  # Limit to first 5 authors

    prompt = f"""EVALUATE THIS RESEARCH PAPER FOR AGI (ARTIFICIAL GENERAL INTELLIGENCE) POTENTIAL

## PAPER DETAILS
**Title:** {title}
**Authors:** {authors_str}
**Abstract:** {abstract}

## EVALUATION TASK
Rate this paper on each AGI parameter below using a 1-10 scale where:
- **1-3:** No/minimal AGI relevance
- **4-6:** Some AGI potential but limited
- **7-8:** Strong AGI contribution
- **9-10:** Exceptional AGI breakthrough

## AGI PARAMETERS TO EVALUATE

1. **Novel Problem Solving (15% weight):** Does this work solve new types of problems not seen in training data or traditional approaches?

2. **Few-Shot Learning (15% weight):** Does the approach learn new tasks or concepts from very few examples?

3. **Task Transfer (15% weight):** Can the learned capabilities transfer across different domains or tasks?

4. **Abstract Reasoning (12% weight):** Does it demonstrate logical reasoning, pattern recognition, or conceptual understanding?

5. **Contextual Adaptation (10% weight):** Does the system adapt its behavior based on different contexts or situations?

6. **Multi-Rule Integration (10% weight):** Can it follow and integrate multiple complex rules or constraints simultaneously?

7. **Generalization Efficiency (8% weight):** How well does it generalize to new scenarios with minimal additional data?

8. **Meta-Learning (8% weight):** Does it demonstrate learning how to learn or adaptive learning strategies?

9. **World Modeling (4% weight):** Does it build or use models of complex environments or systems?

10. **Autonomous Goal Setting (3% weight):** Can it set and pursue its own objectives rather than just following instructions?

## OUTPUT FORMAT
Provide your evaluation as a JSON object with this exact structure:

```json
{{
    "parameter_scores": {{
        "novel_problem_solving": {{"score": X, "reasoning": "explanation"}},
        "few_shot_learning": {{"score": X, "reasoning": "explanation"}},
        "task_transfer": {{"score": X, "reasoning": "explanation"}},
        "abstract_reasoning": {{"score": X, "reasoning": "explanation"}},
        "contextual_adaptation": {{"score": X, "reasoning": "explanation"}},
        "multi_rule_integration": {{"score": X, "reasoning": "explanation"}},
        "generalization_efficiency": {{"score": X, "reasoning": "explanation"}},
        "meta_learning": {{"score": X, "reasoning": "explanation"}},
        "world_modeling": {{"score": X, "reasoning": "explanation"}},
        "autonomous_goal_setting": {{"score": X, "reasoning": "explanation"}}
    }},
    "overall_agi_assessment": "2-3 sentence summary of AGI potential",
    "key_innovations": ["innovation1", "innovation2", "innovation3"],
    "limitations": ["limitation1", "limitation2"],
    "confidence_level": "High/Medium/Low"
}}
```

## EVALUATION GUIDELINES
- Be conservative: Reserve high scores (7+) for truly exceptional AGI contributions
- Focus on what the paper actually demonstrates, not just what it claims
- Consider both theoretical contributions and empirical results
- Distinguish between narrow AI improvements and genuine AGI progress
- Look for evidence of general intelligence capabilities beyond domain-specific performance

Evaluate the paper now:"""

    return prompt

def calculate_agi_score(parameter_scores: Dict[str, float]) -> tuple[float, Dict[str, Any]]:
    """
    Calculate the weighted AGI score from individual parameter scores.

    Args:
        parameter_scores: Dictionary mapping parameter names to scores (1-10)

    Returns:
        Tuple of (weighted_score, score_breakdown_dict)
    """
    total_weighted_score = 0.0
    total_weight = 0.0
    parameter_contributions = {}

    for param_name, param_config in AGI_PARAMETERS.items():
        if param_name in parameter_scores:
            score = parameter_scores[param_name]
            weight = param_config["weight"]

            contribution = score * weight     # Scale 0-10

            total_weighted_score += contribution
            total_weight += weight
            parameter_contributions[param_name] = {
                "score": score,
                "weight": weight,
                "contribution": round(contribution, 1)
            }

    # Normalize to 0-100 scale
    if total_weight > 0:
        final_score = (total_weighted_score / total_weight) * 10      # scale 0-100
    else:
        final_score = 0.0

    # Determine classification
    if final_score >= 70:
        classification = "High AGI Potential"
    elif final_score >= 40:
        classification = "Medium AGI Potential"
    else:
        classification = "Low AGI Potential"

    breakdown = {
        "final_score": round(final_score, 1),
        "classification": classification,
        "parameter_contributions": parameter_contributions,
        "total_weight_used": total_weight
    }

    return round(final_score, 1), breakdown
````

````python
class PlannerAgent:
    """Planner Agent - Creates execution plans for research objectives."""


    def __init__(self):
        self.config = get_config()
        self.logger = get_logger("planner_agent")

        # Initialize OpenAI LLM
        self.llm = ChatOpenAI(model_name="gpt-4o-mini",
            temperature=0.2
        )

        # System prompt for planning
        self.system_prompt = """You are a Planning Specialist in an Artificial General Intelligence research system.

Your goal is to create a comprehensive research execution plan based on the given objective.

TASK: Create a comprehensive research execution plan based on the given objective.

OUTPUT FORMAT: You MUST return a valid JSON object with the following structure:
{
    "search_keywords": ["keyword1", "keyword2", ...],
    "search_strategy": {
        "primary_sources": ["arxiv"],
        "categories": ["cs.AI", "cs.LG", "cs.CL"],
        "date_range": "YYYY-MM-DD to YYYY-MM-DD",
        "max_papers_per_source": 10
    },
    "success_criteria": {
        "min_papers": 10,
        "min_high_agi_papers": 2,
        "quality_threshold": 0.7
    },
    "focus_areas": ["specific area 1", "specific area 2"],
    "exclusions": ["topics to avoid"],
    "special_instructions": "Any specific guidance for downstream agents"
}

PLANNING GUIDELINES:
1. Extract 5-10 highly relevant search keywords
2. Set appropriate time windows (recent work for current trends)
3. Define measurable success criteria
4. Focus on 2-5 specific research areas
5. List topics that should be avoided

Be thorough, specific, and actionable."""

    def plan(self, state: ResearchSystemState) -> Dict[str, Any]:
        """Create an execution plan for the research objective."""

        research_objective = state.get("research_objective", "")
        request_id = state.get("request_id", "")

        self.logger.info(f"Creating execution plan for request {request_id}")

        # Calculate date range (last N days)
        end_date = datetime.now()
        start_date = end_date - timedelta(days=self.config.research_days_lookback)
        date_range = f"{start_date.strftime('%Y-%m-%d')} to {end_date.strftime('%Y-%m-%d')}"

        # Create the planning prompt
        user_prompt = f"""Create a detailed execution plan for the following research objective:

RESEARCH OBJECTIVE: {research_objective}

The current date is: {datetime.now().isoformat()}
Default date range: {date_range}

Additional Context:
- This is for an AGI-focused research system
- We have access to arXiv database
- The system will process papers through discovery, evaluation phases
- Focus on recent, high-impact work in AI/ML

Generate the execution plan now."""

        try:
            # Invoke LLM with system and user prompts
            response = call_llm_with_retry(
                llm = self.llm,
                messages = [
                    SystemMessage(content=self.system_prompt),
                    HumanMessage(content=user_prompt)
                    ],
                logger = self.logger)

            # Parse the JSON response
            plan_content = response.content

            # Extract JSON from response (handle markdown formatting)
            if "```json" in plan_content:
                json_start = plan_content.find("```json") + 7
                json_end = plan_content.find("```", json_start)
                plan_content = plan_content[json_start:json_end]

            execution_plan = json.loads(plan_content.strip())

            self.logger.info(f"Execution Plan : {execution_plan}")

            # Ensure date range is set
            if not execution_plan.get("search_strategy", {}).get("date_range"):
                execution_plan["search_strategy"]["date_range"] = date_range

            self.logger.info(f"Execution plan created successfully",
                           keywords=len(execution_plan.get("search_keywords", [])),
                           focus_areas=len(execution_plan.get("focus_areas", [])))

            return execution_plan

        except json.JSONDecodeError as e:
            self.logger.error(f"Failed to parse plan JSON: {e}")
            # Return a default plan as fallback
            return self._create_default_plan(research_objective, date_range)
        except Exception as e:
            self.logger.error(f"Error creating execution plan: {e}")
            return self._create_default_plan(research_objective, date_range)

    def _create_default_plan(self, objective: str, date_range: str) -> Dict[str, Any]:
        """Create a fallback execution plan."""
        return {
            "search_keywords": ["AGI", "artificial general intelligence", "general AI"],
            "search_strategy": {
                "primary_sources": ["arxiv"],
                "categories": ["cs.AI", "cs.LG"],
                "date_range": date_range,
                "max_papers_per_source": 10
            },
            "success_criteria": {
                "min_papers": 10,
                "min_high_agi_papers": 2
            },
            "focus_areas": ["general artificial intelligence"],
            "exclusions": [],
            "special_instructions": f"Default plan for: {objective}"
        }

def planner_node(state: ResearchSystemState) -> ResearchSystemState:
    """LangGraph node function for the Planner Agent."""

    logger = get_logger("planner_node")
    logger.info("=== PLANNER NODE - START ===")
    logger.info(f"Request ID: {state.get('request_id', 'Not set')}")
    logger.info(f"Research Objective: {state.get('research_objective', 'Not set')}")

    try:
        # Create planner instance
        planner = PlannerAgent()

        # Generate execution plan
        # execution_plan = asyncio.run(planner.plan(state))
        execution_plan = planner.plan(state)

        # Update state with the plan
        if "synthesis_data" not in state or state["synthesis_data"] is None:
            state["synthesis_data"] = {}

        state["synthesis_data"]["execution_plan"] = execution_plan
        state["synthesis_data"]["plan_created_at"] = datetime.now().isoformat()

        logger.info("Execution plan created successfully",
                   keywords=len(execution_plan.get("search_keywords", [])),
                   focus_areas=len(execution_plan.get("focus_areas", [])))

        return state

    except Exception as e:
        logger.error(f"Planner node error: {e}")

        # Add error to state
        if "errors" not in state:
            state["errors"] = []

        state["errors"].append({
            "error_type": type(e).__name__,
            "error_message": str(e),
            "context": {"phase": "planning", "request_id": state.get("request_id")},
            "timestamp": datetime.now().isoformat()
        })

        return state
````

````python
def create_discovery_supervisor() -> Any:
    """Create and return a Discovery Supervisor agent using LangGraph's create_agent."""

    config = get_config()

    # Initialize OpenAI LLM
    llm = ChatOpenAI(model_name="gpt-4o-mini")

    # Create the system prompt for the Discovery Supervisor
    system_prompt = """You are the Discovery Supervisor Agent. Execute paper discovery based on the provided execution plan.

TASK: Follow the execution plan to discover relevant research papers using a streamlined single-tool process.

AVAILABLE TOOL (Only 1 tool):
discover_and_process_papers: Complete discovery workflow tool
   - Parameters:
     * query (str): Search query for papers
     * max_papers (int): Maximum total papers to retrieve (default: 10)
     * from_date (str): Start date (YYYY-MM-DD format, optional)
     * to_date (str): End date (YYYY-MM-DD format, optional)
   - Automatically performs:
     * Searches arXiv source
     * Consolidates all results
     * Deduplication and validation
     * Date filtering
     * Statistics generation

HOW TO USE THE TOOL:

SINGLE STEP - DISCOVER AND PROCESS:
- Call discover_and_process_papers with the execution plan parameters
- For query: Use a simple combination of 2-3 main keywords (e.g., "AGI OR artificial general intelligence")
- Avoid overly complex queries with too many OR/AND/NOT operators

EXECUTION PLAN INTERPRETATION:
- search_keywords: Use 2-3 most relevant keywords for the query (not all 10)
- max_papers_per_source: Use this value directly as max_papers parameter
- date_range: Convert "2025-08-12 to 2025-08-19" to from_date="2025-08-12", to_date="2025-08-19"

IMPORTANT:
- Call the tool ONLY ONCE per execution plan
- The tool returns a complete dictionary with processed_papers, statistics, source_counts, and search_metadata
- Use this returned data as your final output

Keep search queries simple - use 2-3 main keywords, not complex boolean expressions."""

    # Create the agent with tools
    agent = create_agent(
        model=llm,
        tools=DISCOVERY_TOOLS,
        system_prompt=system_prompt
    )

    logger = get_logger("discovery_supervisor")
    logger.info("Discovery Supervisor agent created successfully")
    return agent

def discovery_supervisor_node(state: ResearchSystemState) -> ResearchSystemState:
    """LangGraph node function that executes the Discovery Supervisor agent."""

    logger = get_logger("discovery_supervisor_node")
    logger.info("=== DISCOVERY SUPERVISOR NODE - START ===")
    logger.info(f"Request ID: {state.get('request_id', 'Not set')}")
    logger.info(f"Research Objective: {state.get('research_objective', 'Not set')}")
    logger.info(f"Current Phase: {state.get('current_phase', 'unknown')}")

    try:
        # Create the discovery agent
        agent = create_discovery_supervisor()

        # Prepare the execution plan message
        execution_plan = state.get("synthesis_data", {}).get("execution_plan", {})

        if not execution_plan:
            logger.warning("No execution plan found, using fallback")
            input_message = f"""Research Objective: {state.get('research_objective', 'General AGI research')}

Execute discovery for this objective using discover_and_process_papers tool.
Use simple keywords and keep results to maximum 10 papers total."""
        else:
            logger.info(f"Execution Plan found with {len(execution_plan.get('search_keywords', []))} keywords")

            # Extract date range from execution plan
            date_range = execution_plan.get('search_strategy', {}).get('date_range', '')
            if ' to ' in date_range:
                # Parse "2025-08-12 to 2025-08-19" format
                date_parts = date_range.split(' to ')
                from_date = date_parts[0].strip() if len(date_parts) > 0 else None
                to_date = date_parts[1].strip() if len(date_parts) > 1 else None
            else:
                from_date = None
                to_date = None

            # Create structured input message with execution plan
            input_message = f"""EXECUTION PLAN:

Keywords: {execution_plan.get('search_keywords', [])}
Date Range: {from_date} to {to_date}
Max Papers: {execution_plan.get('search_strategy', {}).get('max_papers_per_source', 10)}

Focus Areas: {execution_plan.get('focus_areas', [])}
Exclusions: {execution_plan.get('exclusions', [])}

Execute discovery using discover_and_process_papers tool with the parameters above.
Use 2-3 main keywords from the list for the query."""

        logger.info(f"Invoking discovery agent with execution plan")

        # Invoke the agent
        result = agent.invoke({
            "messages": [HumanMessage(content=input_message)]
        })

        # Extract structured results from discover_and_process_papers tool call
        discovered_papers = []
        processing_statistics = {}
        source_counts = {}
        search_metadata = {
            "agent_reasoning": [],
            "tool_calls": [],
            "sources_searched": [],
            "processing_time": datetime.now().isoformat()
        }

        # Process agent messages to find discover_and_process_papers result
        logger.info(f"Processing {len(result.get('messages', []))} messages from agent")

        for i, message in enumerate(result.get("messages", [])):
            message_type = type(message).__name__

            # Track agent reasoning
            if hasattr(message, 'content') and message.content:
                search_metadata["agent_reasoning"].append({
                    "type": message_type,
                    "content": message.content[:200] + "..." if len(message.content) > 200 else message.content
                })

            # Extract tool calls
            if hasattr(message, 'tool_calls'):
                for tool_call in message.tool_calls:
                    tool_name = tool_call.get("name", "unknown")
                    search_metadata["tool_calls"].append({
                        "tool": tool_name,
                        "args": tool_call.get("args", {}),
                        "timestamp": datetime.now().isoformat()
                    })

                    if tool_name == "discover_and_process_papers":
                        search_metadata["sources_searched"] = ["arxiv"]

            # Look for discover_and_process_papers tool result
            if hasattr(message, 'content') and message_type == "ToolMessage":
                if isinstance(message.content, str):
                    try:
                        tool_result = json.loads(message.content)

                        # Check if this looks like discover_and_process_papers output
                        if isinstance(tool_result, dict) and 'processed_papers' in tool_result:
                            discovered_papers = tool_result.get('processed_papers', [])
                            processing_statistics = tool_result.get('statistics', {})
                            source_counts = tool_result.get('source_counts', {})

                            # Extract search metadata from the tool result
                            tool_search_metadata = tool_result.get('search_metadata', {})
                            if tool_search_metadata:
                                search_metadata.update(tool_search_metadata)

                            break  # Found the main result, stop processing
                    except json.JSONDecodeError:
                        continue

        # Update search metadata with final counts
        search_metadata["total_found"] = processing_statistics.get('initial_count', len(discovered_papers))
        search_metadata["final_papers"] = len(discovered_papers)
        search_metadata["processing_statistics"] = processing_statistics
        search_metadata["source_counts"] = source_counts

        # Update the research system state with structured data
        state["discovered_papers"] = discovered_papers

        # Store comprehensive metadata in synthesis_data
        if "synthesis_data" not in state:
            state["synthesis_data"] = {}
        state["synthesis_data"]["discovery_metadata"] = search_metadata

        # Log results
        logger.info(f"Discovery completed",
                   papers_discovered=len(discovered_papers),
                   processing_stats=processing_statistics,
                   source_distribution=source_counts)

        if discovered_papers:
            # Show sample papers
            for i, paper in enumerate(discovered_papers[:3], 1):
                title = paper.get('title', 'Unknown Title')
                metadata = paper.get('metadata', {})
                authors = metadata.get('authors', [])
                published_date = metadata.get('published_date', 'Unknown')
                source = metadata.get('source', 'Unknown')

                author_names = []
                if authors and isinstance(authors, list):
                    for author in authors[:2]:
                        if isinstance(author, dict):
                            author_names.append(author.get('name', 'Unknown'))
                        elif isinstance(author, str):
                            author_names.append(author)

                if not author_names:
                    author_names = ['Unknown Author']

                logger.info(f"Sample Paper {i}: {title[:60]}...",
                           authors=', '.join(author_names),
                           published=published_date,
                           source=source)

        return state

    except Exception as e:
        logger.error(f"Discovery Supervisor error: {e}")

        # Add error to state
        error_info = {
            "error_type": type(e).__name__,
            "error_message": str(e),
            "context": {"phase": "discovery", "request_id": state.get("request_id")},
            "timestamp": datetime.now().isoformat()
        }

        if "errors" not in state:
            state["errors"] = []
        state["errors"].append(error_info)

        # Set empty papers list as fallback
        state["discovered_papers"] = []

        return state
````

````python
def _validate_paper_for_evaluation(paper: Dict[str, Any], index: int, request_id: str, logger: ResearchLogger) -> bool:
    """Validate that a paper has the required structure for evaluation."""
    try:
        # Check required fields
        required_fields = ["id", "title", "metadata"]
        missing_fields = [field for field in required_fields if field not in paper]

        if missing_fields:
            logger.warning(f"Paper {index} missing required fields: {missing_fields}")
            return False

        # Check metadata structure
        metadata = paper.get("metadata", {})
        if not isinstance(metadata, dict):
            logger.warning(f"Paper {index} metadata is not a dictionary")
            return False

        # Check for abstract (required for evaluation)
        abstract = metadata.get("abstract", "").strip()
        if not abstract:
            logger.warning(f"Paper {index} has no abstract")
            return False

        # Check abstract length (minimum requirement)
        if len(abstract) < 50:
            logger.warning(f"Paper {index} abstract too short ({len(abstract)} chars)")
            return False

        return True

    except Exception as e:
        logger.error(f"Paper validation error for paper {index}: {e}")
        return False


def generate_paper_analysis_section(paper: Dict[str, Any], index: int) -> str:
    """Generate detailed analysis section for a single paper."""
    title = paper.get('paper_title', 'Unknown Title')
    authors = paper.get('paper_authors', [])
    score = paper.get('agi_score', 0)
    classification = paper.get('agi_classification', 'Unknown')
    assessment = paper.get('overall_assessment', 'No assessment available')
    innovations = paper.get('key_innovations', [])
    limitations = paper.get('limitations', [])

    # Format authors
    author_str = ', '.join(authors[:3]) + ('...' if len(authors) > 3 else '')

    analysis = f"""#### {index}. {title}
**Authors:** {author_str}
**AGI Score:** {score:.1f}/100 ({classification})
**Source:** {paper.get('paper_source', 'unknown')}

**Key Innovations:**
"""

    for innovation in innovations[:3]:
        analysis += f"- {innovation}\n"

    if not innovations:
        analysis += "- No specific innovations identified\n"

    analysis += f"""
**Assessment:** {assessment}

"""

    if limitations:
        analysis += f"**Limitations:**\n"
        for limitation in limitations[:2]:
            analysis += f"- {limitation}\n"

    analysis += "\n---\n\n"
    return analysis


def generate_detailed_evaluation_report(evaluation_results: List[Dict[str, Any]], metadata: Dict[str, Any], request_id: str) -> str:
    """Generate a comprehensive detailed evaluation report."""

    # Handle empty evaluation results case
    if len(evaluation_results) == 0:
        return f"""# Detailed AGI Evaluation Report

**Request ID:** {request_id}
**Generated:** {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}
**Total Papers Evaluated:** 0

---

## Executive Summary

### Overall Statistics
- **Papers Evaluated:** 0
- **Average AGI Score:** 0.0/100
- **High AGI Potential (70):** 0 papers (0.0%)
- **Medium AGI Potential (40-69):** 0 papers (0.0%)
- **Low AGI Potential (<40):** 0 papers (0.0%)

---

## Analysis

No papers were successfully evaluated. This could be due to:
- No papers were discovered during the research phase
- All discovered papers were filtered out due to insufficient content
- Evaluation errors occurred for all papers

## Recommendations

1. **Expand Search Terms**: Try broader or alternative keywords for paper discovery
2. **Check Data Sources**: Ensure the research sources are accessible
3. **Review Filters**: Check if quality filters are too restrictive
4. **Manual Review**: Consider adding papers manually if automatic discovery failed

---

*Report generated by Multi-Agent Research System*
"""

    # Sort papers by AGI score
    sorted_papers = sorted(evaluation_results, key=lambda x: x.get('agi_score', 0), reverse=True)

    # Categorize papers by AGI potential
    high_agi_papers = [p for p in sorted_papers if p.get('agi_score', 0) >= 70]
    medium_agi_papers = [p for p in sorted_papers if 40 <= p.get('agi_score', 0) < 70]
    low_agi_papers = [p for p in sorted_papers if p.get('agi_score', 0) < 40]

    # Generate comprehensive report
    total_papers = len(evaluation_results)
    high_pct = (len(high_agi_papers) / total_papers * 100) if total_papers > 0 else 0
    medium_pct = (len(medium_agi_papers) / total_papers * 100) if total_papers > 0 else 0
    low_pct = (len(low_agi_papers) / total_papers * 100) if total_papers > 0 else 0

    highest_score = max(p.get('agi_score', 0) for p in sorted_papers) if sorted_papers else 0
    lowest_score = min(p.get('agi_score', 0) for p in sorted_papers) if sorted_papers else 0
    score_range = highest_score - lowest_score

    report = f"""# Detailed AGI Evaluation Report

**Request ID:** {request_id}
**Generated:** {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}
**Total Papers Evaluated:** {total_papers}

---

## Executive Summary

### Overall Statistics
- **Papers Evaluated:** {total_papers}
- **Average AGI Score:** {metadata.get('avg_agi_score', 0):.1f}/100
- **High AGI Potential (70):** {len(high_agi_papers)} papers ({high_pct:.1f}%)
- **Medium AGI Potential (40-69):** {len(medium_agi_papers)} papers ({medium_pct:.1f}%)
- **Low AGI Potential (<40):** {len(low_agi_papers)} papers ({low_pct:.1f}%)

### Score Distribution
- **Highest Score:** {highest_score:.1f}/100
- **Lowest Score:** {lowest_score:.1f}/100
- **Score Range:** {score_range:.1f} points

---

## Detailed Paper Analysis

"""

    # Add analysis for each category
    if high_agi_papers:
        report += f"""###  HIGH AGI POTENTIAL PAPERS (70 points)

These papers demonstrate significant contributions to AGI development:

"""
        for i, paper in enumerate(high_agi_papers, 1):
            report += generate_paper_analysis_section(paper, i)

    if medium_agi_papers:
        report += f"""###  MEDIUM AGI POTENTIAL PAPERS (40-69 points)

These papers show promising developments but need stronger AGI relevance:

"""
        for i, paper in enumerate(medium_agi_papers, 1):
            report += generate_paper_analysis_section(paper, i)

    if low_agi_papers:
        report += f"""###  LOW AGI POTENTIAL PAPERS (<40 points)

These papers focus on narrow AI applications with limited AGI relevance:

"""
        for i, paper in enumerate(low_agi_papers, 1):
            report += generate_paper_analysis_section(paper, i)

    # Add recommendations
    report += f"""---

## Recommendations for Future Research

### For Researchers:
"""

    if high_agi_papers:
        report += f"""- **Build on High-Potential Work:** Focus on extending the approaches from the {len(high_agi_papers)} high-scoring papers
- **Address Scalability:** Most high-scoring papers still lack large-scale demonstrations
"""

    if len(low_agi_papers) > len(high_agi_papers):
        report += f"""- **Broaden AGI Focus:** {len(low_agi_papers)} papers focused on narrow applications - consider general intelligence applications
- **Emphasize Transfer Learning:** Many papers lack cross-domain generalization capabilities
- **Include AGI Benchmarks:** Use established AGI evaluation frameworks in your work
"""

    report += f"""
### For This Research System:
- **Search Strategy:** {'Consider broader keywords' if len(low_agi_papers) > len(high_agi_papers) else 'Current search strategy is effective'}
- **Evaluation Criteria:** {'Parameters are appropriately strict' if len(high_agi_papers) < len(evaluation_results) * 0.3 else 'Consider refining AGI relevance criteria'}

---

## Scoring Methodology

### AGI Parameters (Weighted):
1. **Novel Problem Solving** (15% weight) - Creative approaches to unsolved problems
2. **Few-Shot Learning** (15% weight) - Learning from minimal examples
3. **Task Transfer** (15% weight) - Applying skills across domains
4. **Abstract Reasoning** (12% weight) - Logical and conceptual thinking
5. **Contextual Adaptation** (10% weight) - Flexible responses to situations
6. **Multi-Rule Integration** (10% weight) - Combining multiple constraints
7. **Generalization Efficiency** (8% weight) - Broad applicability with minimal data
8. **Meta-Learning** (8% weight) - Learning how to learn
9. **World Modeling** (4% weight) - Understanding complex environments
10. **Autonomous Goal Setting** (3% weight) - Independent objective formation

### Score Interpretation:
- **90-100:** Exceptional AGI contribution (breakthrough level)
- **70-89:** High AGI potential (significant advancement)
- **40-69:** Medium AGI potential (promising development)
- **0-39:** Low AGI potential (narrow AI focus)

---

*This detailed report provides comprehensive insights into the AGI potential of evaluated papers, helping identify promising research directions and common limitations in current AI research.*
"""

    return report
````

````python
def create_evaluation_supervisor() -> ChatOpenAI:
    """Create and return an Evaluation Supervisor LLM instance."""

    config = get_config()

    # Initialize OpenAI LLM
    llm = ChatOpenAI(model_name="gpt-4o-mini")

    logger = get_logger("evaluation_supervisor")
    logger.info("Evaluation Supervisor LLM created successfully")
    return llm

def evaluation_supervisor_node(state: ResearchSystemState) -> ResearchSystemState:
    """LangGraph node function that executes the Evaluation Supervisor."""

    logger = get_logger("evaluation_supervisor_node")
    logger.info("=== EVALUATION SUPERVISOR NODE - START ===")
    logger.info(f"Request ID: {state.get('request_id', 'Not set')}")
    logger.info(f"Research Objective: {state.get('research_objective', 'Not set')}")
    logger.info(f"Discovered Papers: {len(state.get('discovered_papers', []))}")

    try:
        # Get discovered papers from state
        discovered_papers = state.get("discovered_papers", [])
        request_id = state.get("request_id", "unknown")

        if not discovered_papers:
            logger.warning("No discovered papers found in state - nothing to evaluate")
            state["evaluation_results"] = []
            return state

        # Validate papers structure
        valid_papers = []
        for i, paper in enumerate(discovered_papers):
            if _validate_paper_for_evaluation(paper, i, request_id, logger):
                valid_papers.append(paper)

        if not valid_papers:
            logger.warning("No valid papers for evaluation after validation")
            state["evaluation_results"] = []
            return state

        logger.info(f"Starting evaluation of {len(valid_papers)} valid papers")

        # Create the evaluation LLM
        llm = create_evaluation_supervisor()

        # Initialize evaluation metadata
        evaluation_metadata = {
            "total_papers": len(valid_papers),
            "evaluated_papers": 0,
            "successful_evaluations": 0,
            "failed_evaluations": 0,
            "avg_agi_score": 0,
            "score_distribution": {"high": 0, "medium": 0, "low": 0},
            "processing_time": datetime.now().isoformat(),
            "evaluation_errors": []
        }

        evaluation_results = []
        total_agi_score = 0

        # Evaluate each paper
        for i, paper in enumerate(valid_papers, 1):
            logger.info(f"Evaluating Paper {i}/{len(valid_papers)}")

            try:
                # Extract paper details
                title = paper.get('title', 'Unknown Title')
                metadata = paper.get('metadata', {})
                abstract = metadata.get('abstract', '')
                authors = metadata.get('authors', [])

                # Extract author names
                author_names = []
                if isinstance(authors, list):
                    for author in authors:
                        if isinstance(author, dict):
                            author_names.append(author.get('name', 'Unknown'))
                        elif isinstance(author, str):
                            author_names.append(author)

                logger.info(f"Paper {i}: {title[:50]}...",
                           authors=len(author_names),
                           abstract_length=len(abstract))

                if not abstract or len(abstract) < 50:
                    logger.warning(f"Skipping paper {i} - insufficient abstract content")
                    evaluation_metadata["failed_evaluations"] += 1
                    continue

                # Generate AGI evaluation prompt
                agi_prompt = get_agi_metrics_prompt(title, abstract, author_names)

                # Create system message for evaluation
                system_message = SystemMessage(content="""You are an expert AGI evaluator. Analyze research papers for their contribution to Artificial General Intelligence advancement.

Provide precise, evidence-based evaluations in the requested JSON format. Be conservative in scoring - reserve high scores (7-10) for truly exceptional AGI contributions.""")

                human_message = HumanMessage(content=agi_prompt)

                # Get evaluation from LLM
                logger.info(f"Requesting AGI evaluation from LLM for paper {i}")
                response = call_llm_with_retry(
                                                llm = llm,
                                                messages = [system_message, human_message],
                                                logger = logger
                                              )

                # Parse the response
                evaluation_text = response.content

                # Extract JSON from response (handle potential markdown formatting)
                json_start = evaluation_text.find('{')
                json_end = evaluation_text.rfind('}') + 1

                if json_start >= 0 and json_end > json_start:
                    json_content = evaluation_text[json_start:json_end]
                    try:
                        evaluation_data = json.loads(json_content)

                        # Extract parameter scores for weighted calculation
                        parameter_scores = {}
                        if 'parameter_scores' in evaluation_data:
                            for param, details in evaluation_data['parameter_scores'].items():
                                if isinstance(details, dict) and 'score' in details:
                                    parameter_scores[param] = details['score']

                        # Calculate weighted AGI score
                        weighted_score, score_breakdown = calculate_agi_score(parameter_scores)

                        # Create comprehensive evaluation result
                        paper_evaluation = {
                            "paper_id": paper.get('id', f'paper_{i}'),
                            "paper_title": title,
                            "paper_authors": author_names,
                            "paper_source": metadata.get('source', 'unknown'),
                            "paper_url": paper.get('link', ''),
                            "evaluation_timestamp": datetime.now().isoformat(),
                            "agi_score": weighted_score,
                            "agi_classification": score_breakdown["classification"],
                            "parameter_scores": evaluation_data.get('parameter_scores', {}),
                            "overall_assessment": evaluation_data.get('overall_agi_assessment', ''),
                            "key_innovations": evaluation_data.get('key_innovations', []),
                            "limitations": evaluation_data.get('limitations', []),
                            "confidence_level": evaluation_data.get('confidence_level', 'Medium'),
                            "score_breakdown": score_breakdown
                        }

                        evaluation_results.append(paper_evaluation)
                        total_agi_score += weighted_score
                        evaluation_metadata["successful_evaluations"] += 1

                        # Update score distribution
                        classification = score_breakdown["classification"]
                        if "High" in classification:
                            evaluation_metadata["score_distribution"]["high"] += 1
                        elif "Medium" in classification:
                            evaluation_metadata["score_distribution"]["medium"] += 1
                        else:
                            evaluation_metadata["score_distribution"]["low"] += 1

                        logger.info(f"Paper {i} evaluated - AGI Score: {weighted_score}/100 ({classification})")

                    except json.JSONDecodeError as e:
                        logger.error(f"Failed to parse LLM evaluation JSON for paper {i}: {e}")
                        evaluation_metadata["failed_evaluations"] += 1
                        evaluation_metadata["evaluation_errors"].append(f"Paper {i}: JSON parse error")

                else:
                    logger.error(f"No valid JSON found in LLM response for paper {i}")
                    evaluation_metadata["failed_evaluations"] += 1
                    evaluation_metadata["evaluation_errors"].append(f"Paper {i}: No JSON in response")

                evaluation_metadata["evaluated_papers"] += 1

            except Exception as e:
                logger.error(f"Error evaluating paper {i}: {e}")
                evaluation_metadata["failed_evaluations"] += 1
                evaluation_metadata["evaluation_errors"].append(f"Paper {i}: {str(e)}")

        # Calculate final statistics
        if evaluation_metadata["successful_evaluations"] > 0:
            evaluation_metadata["avg_agi_score"] = round(
                total_agi_score / evaluation_metadata["successful_evaluations"], 1
            )

        # Update state with evaluation results
        state["evaluation_results"] = evaluation_results

        # Store evaluation metadata in synthesis_data
        if "synthesis_data" not in state or state["synthesis_data"] is None:
            state["synthesis_data"] = {}
        state["synthesis_data"]["evaluation_metadata"] = evaluation_metadata

        # Log results summary
        logger.info("Evaluation completed",
                   total_papers=evaluation_metadata['total_papers'],
                   successful_evaluations=evaluation_metadata['successful_evaluations'],
                   failed_evaluations=evaluation_metadata['failed_evaluations'],
                   avg_agi_score=evaluation_metadata['avg_agi_score'],
                   score_distribution=evaluation_metadata['score_distribution'])

        if evaluation_results:
            # Show top papers
            sorted_papers = sorted(evaluation_results, key=lambda x: x['agi_score'], reverse=True)
            for i, paper in enumerate(sorted_papers[:3], 1):
                logger.info(f"Top Paper {i}: {paper['paper_title'][:50]}...",
                           agi_score=f"{paper['agi_score']}/100",
                           classification=paper['agi_classification'])

        # Generate detailed evaluation report
        try:
            detailed_report = generate_detailed_evaluation_report(
                evaluation_results,
                evaluation_metadata,
                state.get("request_id", "unknown")
            )
            report_filename = f"evaluation_detailed_report_{state.get('request_id', 'unknown')}.md"
            with open(report_filename, "w", encoding="utf-8") as f:
                f.write(detailed_report)
            logger.info(f"Detailed evaluation report written to {report_filename}")
        except Exception as report_error:
            logger.error(f"Failed to generate detailed evaluation report: {report_error}")

        return state

    except Exception as e:
        logger.error(f"Evaluation Supervisor error: {e}")

        # Add error to state
        error_info = {
            "error_type": type(e).__name__,
            "error_message": str(e),
            "context": {"phase": "evaluation", "request_id": state.get("request_id")},
            "timestamp": datetime.now().isoformat()
        }

        if "errors" not in state:
            state["errors"] = []
        state["errors"].append(error_info)

        # Set empty evaluation results as fallback
        state["evaluation_results"] = []

        return state
````

````python
class LeadResearchAgent:
    """Lead Research Agent - Master coordinator for the research system."""

    def __init__(self):
        # self.config = get_config()
        self.logger = get_logger("lead_research_agent")

    def _validate_paper_structure(self, paper: Dict[str, Any], index: int) -> bool:
        """Validate that a paper has the required structure."""
        try:
            # Check required fields
            required_fields = ["id", "title", "metadata"]
            missing_fields = [field for field in required_fields if field not in paper]

            if missing_fields:
                self.logger.warning(f"Paper {index} missing required fields: {missing_fields}")
                return False

            # Check metadata structure
            metadata = paper.get("metadata", {})
            if not isinstance(metadata, dict):
                self.logger.warning(f"Paper {index} metadata is not a dictionary")
                return False

            # Check for abstract (required for evaluation)
            if not metadata.get("abstract"):
                self.logger.warning(f"Paper {index} has no abstract")
                return False

            return True

        except Exception as e:
            self.logger.error(f"Paper validation error for paper {index}: {e}")
            return False



    def _initialize(self, state: ResearchSystemState, request: Dict[str, Any]) -> Dict[str, Any]:
        """Initialize research and transition to planning phase."""

        self.logger.info("Initializing research request")

        # If request_id is empty, create one
        if not state.get("request_id"):
            request_id = str(uuid.uuid4())
            research_objective = request.get("research_objective", state.get("research_objective", ""))
        else:
            request_id = state["request_id"]
            research_objective = state["research_objective"]

        self.logger.info("Phase transition: INITIALIZATION  PLANNING")

        # Return state updates
        updates = {
            "request_id": request_id,
            "research_objective": research_objective,
            "current_phase": ResearchPhase.PLANNING,
            "phase_history": [{
                "from_phase": ResearchPhase.INITIALIZATION,
                "to_phase": ResearchPhase.PLANNING,
                "timestamp": datetime.now()
            }],
            "created_at": datetime.now(),
            "updated_at": datetime.now()
        }

        return {k: v for k, v in updates.items() if v is not None}



    def _generate_final_report_from_evaluation(self, state: ResearchSystemState) -> str:
        """Generate final research report from evaluation results."""

        evaluation_results = state.get("evaluation_results", [])
        discovered_papers = state.get("discovered_papers", [])
        evaluation_metadata = state.get("synthesis_data", {}).get("evaluation_metadata", {})

        if not evaluation_results:
            return "No evaluation results available to generate report."

        # Calculate summary statistics
        total_papers = len(evaluation_results)
        avg_agi_score = evaluation_metadata.get("avg_agi_score", 0)
        score_distribution = evaluation_metadata.get("score_distribution", {})

        # Sort papers by AGI score
        sorted_papers = sorted(evaluation_results, key=lambda x: x.get('agi_score', 0), reverse=True)

        # Create comprehensive report
        report = f"""# AGI Research Analysis Report

## Executive Summary

**Research Objective:** {state.get('research_objective', 'Not specified')}

**Discovery Overview:**
- Total papers discovered: {len(discovered_papers)}
- Papers successfully evaluated: {total_papers}
- Average AGI score: {avg_agi_score:.1f}/100

**AGI Potential Distribution:**
- High AGI Potential: {score_distribution.get('high', 0)} papers
- Medium AGI Potential: {score_distribution.get('medium', 0)} papers
- Low AGI Potential: {score_distribution.get('low', 0)} papers

## Key Findings

### Top-Scoring Papers (AGI Potential)

"""

        # Add top 5 papers with highest AGI scores
        for i, paper in enumerate(sorted_papers[:5], 1):
            title = paper.get('paper_title', 'Unknown Title')
            authors = paper.get('paper_authors', [])
            agi_score = paper.get('agi_score', 0)
            classification = paper.get('agi_classification', 'Unknown')
            key_innovations = paper.get('key_innovations', [])
            overall_assessment = paper.get('overall_assessment', '')

            report += f"""#### {i}. {title}
**Authors:** {', '.join(authors[:3])}{'...' if len(authors) > 3 else ''}
**AGI Score:** {agi_score}/100 ({classification})
**Key Innovations:** {', '.join(key_innovations[:3])}
**Assessment:** {overall_assessment}

"""

        # Add insights and recommendations
        high_potential_papers = len([p for p in evaluation_results if 'High' in p.get('agi_classification', '')])
        medium_potential_papers = len([p for p in evaluation_results if 'Medium' in p.get('agi_classification', '')])

        report += f"""
## Insights and Recommendations

### Current State of AGI Research

Based on the evaluation of {total_papers} papers, the field shows:

"""

        if avg_agi_score >= 70:
            report += "- **Strong AGI advancement:** Multiple papers demonstrate significant progress toward AGI capabilities\n"
        elif avg_agi_score >= 40:
            report += "- **Moderate AGI progress:** Papers show promising developments but need more breakthrough innovations\n"
        else:
            report += "- **Early-stage AGI research:** Most papers focus on narrow AI applications with limited AGI relevance\n"

        if high_potential_papers > 0:
            report += f"- **Breakthrough potential:** {high_potential_papers} papers show high AGI potential\n"
        elif medium_potential_papers > 0:
            report += f"- **Emerging potential:** {medium_potential_papers} papers show medium AGI potential\n"
        else:
            report += f"- **Limited AGI relevance:** Current research focuses on domain-specific applications rather than general intelligence\n"

        report += f"""
### Future Research Directions

"""

        if high_potential_papers > 0:
            report += """1. **Priority Focus:** Deepen analysis of high-potential AGI papers
2. **Collaboration:** Connect with research teams behind breakthrough papers
3. **Follow-up Studies:** Track developments from top-scoring research groups
4. **Implementation:** Consider practical applications of high-scoring innovations
"""
        else:
            report += """1. **Broaden Search:** Expand search criteria to find more AGI-relevant papers
2. **Alternative Sources:** Include additional academic databases and preprint servers
3. **Keyword Enhancement:** Refine search keywords to capture emerging AGI concepts
4. **Timeline Adjustment:** Consider extending the search timeframe
5. **Expert Consultation:** Seek input from AGI researchers on current promising directions
"""

        report += f"""
---
*Report generated on {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}*
*Total processing time: {evaluation_metadata.get('processing_time', 'Unknown')}*
"""

        return report


    def process_phase(self, state: ResearchSystemState, request: Dict[str, Any] = None) -> Dict[str, Any]:
        """Process the current phase and return state updates."""

        current_phase = state["current_phase"]
        request_id = state["request_id"]

        try:
            # Handle each phase appropriately
            if current_phase == ResearchPhase.INITIALIZATION:
                return  self._initialize(state, request)
            elif current_phase == ResearchPhase.PLANNING:
                return  self._post_planning_coordination(state)
            elif current_phase == ResearchPhase.DISCOVERY:
                return  self._post_discovery_coordination(state)
            elif current_phase == ResearchPhase.EVALUATION:
                return  self._post_evaluation_coordination(state)
            else:
                # Already complete or unknown phase
                self.logger.info("Phase processing complete", phase=current_phase.value)
                return {}

        except Exception as e:
            self.logger.error(f"Error in phase processing: {e}")
            return {
                "errors": state.get("errors", []) + [{
                    "error_type": type(e).__name__,
                    "error_message": str(e),
                    "context": {"phase": current_phase.value},
                    "timestamp": datetime.now()
                }]
            }



    def _post_planning_coordination(self, state: ResearchSystemState) -> Dict[str, Any]:
        """Coordinate after planning phase completes."""

        # Check if we have an execution plan
        if state.get("synthesis_data", {}).get("execution_plan"):
            self.logger.info("Phase transition: PLANNING  DISCOVERY")

            plan = state["synthesis_data"]["execution_plan"]
            self.logger.info("Execution plan validated",
                           keywords=len(plan.get("search_keywords", [])))

            return {
                "current_phase": ResearchPhase.DISCOVERY,
                "phase_history": state.get("phase_history", []) + [{
                    "from_phase": ResearchPhase.PLANNING,
                    "to_phase": ResearchPhase.DISCOVERY,
                    "timestamp": datetime.now()
                }],
                "updated_at": datetime.now()
            }
        else:
            # No plan created, move to completion
            self.logger.warning("No execution plan created - moving to completion")
            return {
                "current_phase": ResearchPhase.COMPLETION,
                "final_report": "Failed to create execution plan for the research objective.",
                "phase_history": state.get("phase_history", []) + [{
                    "from_phase": ResearchPhase.PLANNING,
                    "to_phase": ResearchPhase.COMPLETION,
                    "timestamp": datetime.now()
                }],
                "updated_at": datetime.now()
            }

    def _post_discovery_coordination(self, state: ResearchSystemState) -> Dict[str, Any]:
        """Coordinate after discovery phase completes."""

        discovered_papers = state.get("discovered_papers", [])

        self.logger.info("Discovery phase coordination",
                        papers_discovered=len(discovered_papers))

        # Validate discovered papers structure
        valid_papers = []
        if discovered_papers:
            for i, paper in enumerate(discovered_papers):
                if self._validate_paper_structure(paper, i):
                    valid_papers.append(paper)

            if valid_papers:
                # Move to evaluation phase
                self.logger.info("Phase transition: DISCOVERY  EVALUATION",
                               valid_papers=len(valid_papers))

                return {
                    "current_phase": ResearchPhase.EVALUATION,
                    "discovered_papers": valid_papers,
                    "phase_history": state.get("phase_history", []) + [{
                        "from_phase": ResearchPhase.DISCOVERY,
                        "to_phase": ResearchPhase.EVALUATION,
                        "timestamp": datetime.now(),
                        "papers_count": len(valid_papers)
                    }],
                    "updated_at": datetime.now()
                }
            else:
                # No valid papers found
                self.logger.warning("No valid papers found - moving to completion")
                return {
                    "current_phase": ResearchPhase.COMPLETION,
                    "final_report": f"No valid papers were found. {len(discovered_papers)} papers were discovered but failed validation.",
                    "phase_history": state.get("phase_history", []) + [{
                        "from_phase": ResearchPhase.DISCOVERY,
                        "to_phase": ResearchPhase.COMPLETION,
                        "timestamp": datetime.now(),
                        "reason": "no_valid_papers"
                    }],
                    "updated_at": datetime.now()
                }
        else:
            # No papers found at all
            self.logger.warning("No papers discovered - moving to completion")
            return {
                "current_phase": ResearchPhase.COMPLETION,
                "final_report": "No papers were discovered for the given research objective.",
                "phase_history": state.get("phase_history", []) + [{
                    "from_phase": ResearchPhase.DISCOVERY,
                    "to_phase": ResearchPhase.COMPLETION,
                    "timestamp": datetime.now(),
                    "reason": "no_papers_discovered"
                }],
                "updated_at": datetime.now()
            }

    def _post_evaluation_coordination(self, state: ResearchSystemState) -> Dict[str, Any]:
        """Coordinate after evaluation phase completes."""

        evaluation_results = state.get("evaluation_results", [])

        self.logger.info("Evaluation phase coordination",
                        evaluation_results=len(evaluation_results))

        if evaluation_results:
            # Generate final report from evaluation results
            self.logger.info("Phase transition: EVALUATION  COMPLETION")
            final_report = self._generate_final_report_from_evaluation(state)

            # Write report to file
            with open("final_report.md", "w") as f:
                f.write(final_report)

            return {
                "current_phase": ResearchPhase.COMPLETION,
                "final_report": final_report,
                "phase_history": state.get("phase_history", []) + [{
                    "from_phase": ResearchPhase.EVALUATION,
                    "to_phase": ResearchPhase.COMPLETION,
                    "timestamp": datetime.now()
                }],
                "updated_at": datetime.now()
            }
        else:
            # No evaluation results, create summary report
            self.logger.warning("No evaluation results - creating summary report")
            no_results_report = f"""# AGI Research Analysis Report

## Executive Summary

**Research Objective:** {state.get('research_objective', 'Not specified')}

**Discovery Overview:**
- Total papers discovered: {len(state.get('discovered_papers', []))}
- Papers successfully evaluated: 0
- Average AGI score: N/A

## Key Findings

No papers were successfully evaluated for AGI potential.

## Recommendations

1. **Broaden Search Criteria:** Expand search terms and sources
2. **Adjust Date Range:** Consider extending the search timeframe
3. **Review Evaluation Criteria:** Check if evaluation parameters are too restrictive

---
*Report generated on {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}*
"""
            with open("final_report.md", "w") as f:
                f.write(no_results_report)

            return {
                "current_phase": ResearchPhase.COMPLETION,
                "final_report": no_results_report,
                "phase_history": state.get("phase_history", []) + [{
                    "from_phase": ResearchPhase.EVALUATION,
                    "to_phase": ResearchPhase.COMPLETION,
                    "timestamp": datetime.now()
                }],
                "updated_at": datetime.now()
            }


def lead_supervisor_node(state: ResearchSystemState) -> ResearchSystemState:
    """Lead supervisor node for the graph."""

    logger = get_logger("lead_supervisor_node")
    current_phase = state.get("current_phase", ResearchPhase.INITIALIZATION)

    logger.info("=== LEAD SUPERVISOR NODE - START ===")
    logger.info(f"Current Phase: {current_phase}")
    logger.info(f"Request ID: {state.get('request_id', 'Not set')}")
    logger.info(f"Research Objective: {state.get('research_objective', 'Not set')}")

    lead_agent = LeadResearchAgent()

    # Extract the research request from messages (only needed for initialization)
    request = None
    if state.get("current_phase") == ResearchPhase.INITIALIZATION or not state.get("current_phase"):
        messages = state.get("messages", [])
        if messages:
            last_message = messages[-1]
            if hasattr(last_message, 'content'):
                try:
                    request = json.loads(last_message.content)
                except:
                    request = {"research_objective": last_message.content}
            else:
                request = {"research_objective": "General AGI research"}
        else:
            # If no messages but we have research_objective in state, use it
            if state.get("research_objective"):
                request = {"research_objective": state["research_objective"]}
            else:
                request = {"research_objective": "General AGI research"}

    # Process the current phase and get state updates
    state_updates = lead_agent.process_phase(state, request)

    # Log updates
    logger.info("State updates received from lead agent",
               updates=list(state_updates.keys()))

    # Merge updates into state (following LangGraph pattern)
    for key, value in state_updates.items():
        if key == "phase_history" and key in state:
            # Append to existing history
            state[key] = state[key] + value
        elif key == "errors" and key in state:
            # Append to existing errors
            state[key] = state[key] + value
        else:
            # Update or add the key
            state[key] = value

    logger.info(f"Lead supervisor processing complete - New Phase: {state.get('current_phase', 'unknown')}")

    return state
````

````python
def route_next_phase(state: ResearchSystemState) -> Literal["planner", "discovery", "evaluation", "complete"]:
    """Route to the next phase based on current state."""

    current_phase = state.get("current_phase", ResearchPhase.INITIALIZATION)
    request_id = state.get("request_id", "unknown")

    logger = get_logger("routing")
    logger.info("Routing decision",
               request_id=request_id,
               current_phase=str(current_phase),
               discovered_papers_count=len(state.get("discovered_papers", [])),
               evaluation_results_count=len(state.get("evaluation_results", [])))

    # Route based on current phase
    if current_phase == ResearchPhase.PLANNING:
        logger.info("Routing to planner")
        return "planner"
    elif current_phase == ResearchPhase.DISCOVERY:
        logger.info("Routing to discovery")
        return "discovery"
    elif current_phase == ResearchPhase.EVALUATION:
        logger.info("Routing to evaluation")
        return "evaluation"
    elif current_phase == ResearchPhase.COMPLETION:
        logger.info("Routing to completion")
        return "complete"
    else:
        logger.info("Routing to completion (fallback)")
        return "complete"

def route_after_supervisor(state: ResearchSystemState) -> Literal["lead", "complete"]:
    """Route after a supervisor completes its work."""

    if state.get("current_phase") == ResearchPhase.COMPLETION:
        return "complete"
    else:
        return "lead"


def build_research_graph() -> StateGraph:
    """Build the complete research system graph following LangGraph patterns."""

    logger = get_logger("graph_builder")
    logger.info("Building research system graph")

    # Create the main graph
    graph = StateGraph(ResearchSystemState)

    # Add nodes
    graph.add_node("lead_supervisor", lead_supervisor_node)
    graph.add_node("planner", planner_node)
    graph.add_node("discovery_supervisor", discovery_supervisor_node)
    graph.add_node("evaluation_supervisor", evaluation_supervisor_node)

    # Set entry point - always start with lead supervisor
    graph.add_edge(START, "lead_supervisor")

    # Add conditional routing from lead supervisor to phase-specific supervisors
    graph.add_conditional_edges(
        "lead_supervisor",
        route_next_phase,
        {
            "planner": "planner",
            "discovery": "discovery_supervisor",
            "evaluation": "evaluation_supervisor",
            "complete": END
        }
    )

    # Route planner back to lead for coordination
    graph.add_conditional_edges(
        "planner",
        route_after_supervisor,
        {
            "lead": "lead_supervisor",
            "complete": END
        }
    )

    # Route discovery supervisor back to lead for coordination
    graph.add_conditional_edges(
        "discovery_supervisor",
        route_after_supervisor,
        {
            "lead": "lead_supervisor",
            "complete": END
        }
    )

    # Route evaluation supervisor back to lead for coordination
    graph.add_conditional_edges(
        "evaluation_supervisor",
        route_after_supervisor,
        {
            "lead": "lead_supervisor",
            "complete": END
        }
    )

    logger.info("Research system graph built successfully")
    return graph

def create_research_system():
    """Create and compile the complete research system."""

    graph = build_research_graph()
    compiled_graph = graph.compile()

    logger = get_logger("system_creator")
    logger.info("Research system compiled successfully")

    return compiled_graph

def run_research_request(research_objective: str, **kwargs) -> Dict[str, Any]:
    """Run a research request through the system."""

    logger = get_logger("research_runner")
    logger.info(f"research objective : {research_objective}")

    # Generate request ID for tracking
    request_id = str(uuid.uuid4())

    logger.info("Starting research request",
               request_id=request_id,
               objective=research_objective[:100] + "..." if len(research_objective) > 100 else research_objective)

    # Create the system
    try:
        research_system = create_research_system()
        logger.info("Research system created successfully")
    except Exception as e:
        logger.error(f"Failed to create research system: {e}")
        raise

    # Create request message
    request = {
        "research_objective": research_objective,
        **kwargs
    }

    # Initial state
    initial_state = {
        "messages": [HumanMessage(content=str(request))],
        "request_id": request_id,
        "research_objective": research_objective,
        "current_phase": ResearchPhase.INITIALIZATION,
        "phase_history": [],
        "discovered_papers": [],
        "analysis_results": [],
        "evaluation_results": [],
        "synthesis_data": {},
        "final_report": None,
        "pending_tasks": [],
        "completed_tasks": [],
        "errors": [],
        "warnings": [],
        "total_papers_processed": 0,
        "total_processing_time": 0.0
    }

    logger.info("Initial state created", initial_phase=initial_state["current_phase"])

    # Run the system
    try:
        logger.info("Invoking research system")
        start_time = time.time()
        result = research_system.invoke(initial_state)
        processing_time = time.time() - start_time

        # Update processing time in result
        result["total_processing_time"] = processing_time

        # Log completion details
        logger.info("Research request completed successfully",
                   request_id=result.get("request_id"),
                   final_phase=result.get("current_phase"),
                   papers_discovered=len(result.get("discovered_papers", [])),
                   papers_evaluated=len(result.get("evaluation_results", [])),
                   processing_time=f"{processing_time:.1f}s",
                   has_final_report=bool(result.get("final_report")))

        return result

    except Exception as e:
        logger.error(f"Research request failed: {e}")

        # Try to capture partial state for debugging
        try:
            e.partial_state = initial_state
        except:
            pass

        raise
````
